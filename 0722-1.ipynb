```python
filename_data = response.headers['Content-Disposition']
filename =  filename_data.split(";")[-1].split("=")[-1].strip('"')
print(filename)
```

```python
file = open(filename, mode="w",encoding = 'utf-8',newline = "")
file.writelines(response.text)
file.close()
```
